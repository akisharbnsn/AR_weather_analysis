```python
import pandas as pd 
import numpy as np

import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# import data
coffee = pd.read_csv("../data/crop/coffee_output.csv")
```

```python
# TODO: view first 5 rows of your data
coffee.head()
```

```python
# TODO: describe your numeric columns
coffee.describe()
```

```python
# TODO: display the null values in your dataframe
print(coffee.isnull().sum())
```

```python
# TODO: drop your null values
coffee.dropna(inplace=True)
coffee.isnull().sum()
```

```python
# TODO: Select only Minas Gerais

mg_subdiv = coffee[coffee['subdivision'] == 'Minas Gerais']
mg_subdiv['year'] = mg_subdiv['year'].astype(int)
mg_subdiv.to_csv('mgcoffee.csv', index=False)
print(mg_subdiv)
```

# EDA BEGINS HERE!



```python

# Calculate the average (mean) of bearing trees column
mean = coffee['bearing_trees'].mean()

# Calculate the standard deviation of bearing trees column
std = coffee['bearing_trees'].std()

# Plot a histogram of bearing trees with a KDE (Kernel Density Estimation) and an explanatory statement
sns.histplot(mg_subdiv['bearing_trees'], kde=True)
plt.title('Bearing Trees Distribution')
plt.grid(True)
print(f"The average bearing trees distribution is {mean:.2f} with a standard deviation of {std:.2f}.")
```

```python
# Plot a line graph of the number of bearing trees over the years in Minas Gerais
ax = mg_subdiv.plot(x='year', 
               y='bearing_trees', 
               marker='o', 
               title='Bearing Trees Over Time',
               figsize=(12,5))
plt.ylabel('Bearing Trees')
ax.set_xticks(mg_subdiv['year']) 
plt.grid(True)
plt.show()
```

How is bearing trees changing across the years for Minas Gerais? Is this increasing or decreasing? What is the distribution of this variable? Is it normal?

Answer: The number of bearing trees in Minas Gerais depicts an increasing trend as the years progress, which is concluded based on the upward pattern of the line plot. The distribution of the bearing trees is left skewed, meaning it is a not a normal distribution. Most of the data points rest on the higher end of the graph, with some data seen on the left end, slightly bringing down the average. As a result the average distribution of bearing trees is 5646.93, with a standard deviation of 372.68. This data suggests that more trees are maturing and producing fruit, therefore, increasing the amount of coffee beans being produced. 

```python
# Calculate the average (mean) of bearing hectares column
mean = coffee['bearing_hectares'].mean()

# Calculate the standard deviation of bearing hectares column
std = coffee['bearing_hectares'].std()

# Plot and print a histogram of bearing hectares with a KDE and an explanatory statement
sns.histplot(mg_subdiv['bearing_hectares'], kde=True)
plt.title('Bearing Hectares Distribution')
plt.grid(True)
print(f"The average tree bearing hectares distribution is {mean:.2f} with a standard deviation of {std:.2f}.")
```

```python
# Plot a line graph of the number of bearing hectares over the years in Minas Gerais
ax = mg_subdiv.plot(x='year', 
               y='bearing_hectares', 
               marker='o', 
               title='Bearing Hectares Over Time',
               figsize=(12,5))
plt.ylabel('Bearing Hectares')
ax.set_xticks(mg_subdiv['year']) 
plt.grid(True)
plt.show()
```

```python
# Calculate the average (mean) of the 60kgs bags column
mean = coffee['60kgs_bag'].mean()

# Calculate the standard deviation of the 60kgs bags column
std = coffee['60kgs_bag'].std()

# Plot and print a histogram of the 60kgs bags with a KDE and an explanatory statement
sns.histplot(mg_subdiv['60kgs_bag'], kde=True)
plt.title('60kg Coffee Bags Distribution')
plt.grid(True)
print(f"The average 60kg coffee bag distribution is {mean:.2f} with a standard deviation of {std:.2f}.")
```

How is the count of million 60kgs bags of coffee changing across the years for Minas Gerais? Is this increasing or decreasing? What is the distribution of this variable? Is it normal?

Answer: The count of million 60kg bags of coffee in Minas Gerais shows a fluctuation throughout the years, often increasing and decreasing at similar rates. The distribution of this variable appears to be the closest to a normal distribution (a bell shape) versus the other distributions examined.
```python
# Calculate the Pearson correlation between bearing trees and 60kgs bags
correlation = mg_subdiv['bearing_trees'].corr(mg_subdiv['60kgs_bag'])

# Print the correlation value rounded to two decimal places
print(f"The correlation between bearing trees vs 60kgs coffee bags is {correlation:.2f}")
```

```python
# Create a scatter plot with a regression line showing the relationship between the number of bearing trees and 60kgs bags
sns.regplot(x='bearing_trees', 
            y='60kgs_bag', 
            data=mg_subdiv)
plt.grid(True)
plt.show()
```

What is the relationship between bearing_trees and 60kgs_bag? How *strong* is this relationship? Use pearson's correlation coefficient.

Answer: The relationship between bearing_trees and 60kgs_bag is a moderately strong positive correlation, with the Pearson's correlation being 0.66. This leads me to conclude that as the number of trees increase, so do the bags of coffee.
