```python
import pandas as pd 
import numpy as np

import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# import data
weather1 = pd.read_csv("../data/weather/weather_data1.csv")
weather2 = pd.read_csv("../data/weather/weather_data2.csv")
```

```python
# TODO: concatenate your data files together and save to a new csv file ("../data/weather/weather_data.csv")
weather = pd.concat([weather1, weather2], axis=0)
weather.to_csv('weather.csv', index=False)
```

# EDA BEGINS HERE!



```python
# Calculate the average (mean) of maximum rainfall column
mean = weather['rain_max'].mean()

# Calculate the standard deviation of maximum rainfall column
std = weather['rain_max'].std()

# Plot a histogram of maximum rainfall with a KDE (Kernel Density Estimation) and an explanatory statement
sns.histplot(weather['rain_max'], kde=True)
plt.grid(True)
plt.title('Maximum Rain Distribution')
print(f"The average rain distribution is {mean:.2f} with a standard deviation of {std:.2f}.")
```

```python
# Locate the row where 'rain_max' is the highest
max_rain_row = weather.loc[weather['rain_max'].idxmax()]

# Provide a quick summary
print(f"Year with highest rainfall: {max_rain_row['year']}, Rain: {max_rain_row['rain_max']}")
```

Which year had the highest maximum record rainfall?

Answer: 2006 had the highest rainfall with the value being 6.67.

```python
# Calculate the average (mean) of average temperature column
mean = weather['temp_avg'].mean()

# Calculate the standard deviation of average temperature column
std = weather['temp_avg'].std()

# Plot and print a histogram of average temperature with a KDE and an explanatory statement
sns.histplot(weather['temp_avg'], kde=True)
plt.grid(True)
plt.title('Avg Temp Distribution')
print(f"The average temperature is {mean:.2f} degrees celsius with a standard deviation of {std:.2f}.")
```

```python
# Convert the 'year' column to integers for consistent plotting
weather['year'] = weather['year'].astype(int)

# Plot a line graph of average temperature over the years in Minas Gerais
ax = weather.plot(x='year', 
                  y='temp_avg', 
                  kind='line', 
                  marker='o', 
                  title='Avg Temperature Over Time (Minas Gerais)',
                  figsize=(12,5))
plt.ylabel('Average Temperature')
ax.set_xticks(weather['year'])
plt.grid(True) 
plt.show()
```

How is average temperature changing across the years for Minas Gerais? Is this increasing or decreasing?

Answer: The average temperature changes across the years for Minas Gerais showing a drastic drop and decrease in the earlier years, then transitioning to a graudally increase in the most recent years. Using the chart, the evidence depicts the greatest drop in the year 2008, then a gradual uptick from 2008 to 2010, then dropping again in 2011. From 2011 to 2016 we see more consistency in the rate of increase, a slight decrease in 2017 and the pattern repeats, most trending upwards. In summary, it's safe to conclude that the average temperature appears to be gradually increasing over the years in Minas Gerais.

```python
# Calculate the average (mean) of average wind column
mean = weather['wind_avg'].mean()

# Calculare the standard deviation of the average wind column
std = weather['wind_avg'].std()

# Plot and print a histogram of average wind speed with a KDE and an explanatory statement
sns.histplot(weather['wind_avg'], kde=True)
plt.grid(True)
plt.title('Avg Wind Speed')
print(f"The average wind speed is {mean:.2f} with a standard deviation of {std:.2f}.")
```

```python
# Calculate the Pearson correlation between maximum rain and maximum humidity
correlation = weather[['rain_max','hum_max']].corr()
correlation
```

```python
# Print the correlation value rounded to two decimal places
print(f"The correlation between maximum rain and maximum humidity is {correlation:.2f}")
```

```python
# Create a scatter plot with a regression line showing the relationship between maximum rain and maximum humidity
sns.regplot(x='rain_max', 
            y='hum_max',
            data=weather)
plt.grid(True)
plt.show()
```

What is the relationship between max rain and max humidity? How *strong* is this relationship? Use pearson's correlation coefficient.

Answer: There is a moderate/ slightly positive relationship between maximum rainfall and maximum humidity at 0.41. Since Pearson correlation measures how strongly two variables move together, it's safe to infer that based on the visualization this is true. This means that while maximum humidity increases, maximum rainfall tends to increase too, but not too strongly or consistently.
